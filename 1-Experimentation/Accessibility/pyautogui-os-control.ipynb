```python
import pyautogui
import os
import platform
import time
```

```python
def system_info():
    print("System:", platform.system())
    print("OS Version:", platform.version())
    print("Screen size:", pyautogui.size())
```

```python
def move_cursor(x, y, duration=0.2):
    """Move cursor to a position (x, y)."""
    pyautogui.moveTo(x, y, duration=duration)

def click(x=None, y=None, button='left'):
    """Click at current or given position."""
    pyautogui.click(x=x, y=y, button=button)

def type_text(text):
    """Simulate typing."""
    pyautogui.typewrite(text, interval=0.05)

def press_key(key):
    """Press a key (e.g., 'enter', 'ctrl', 'a')."""
    pyautogui.press(key)
```
