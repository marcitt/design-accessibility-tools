```python
import pandas as pd

# Mouse/keyboard
df_mouse = pd.read_csv('ELAN_outputs/bench_1_mouse_keyboard.txt', sep='\t', header=None,
                 names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_mouse = df_mouse[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_mouse = df_mouse.groupby('value')['dur_s'].mean().reset_index()
result_mouse['task_num'] = result_mouse['value'].str.extract(r'(\d+)').astype(int)
result_mouse = result_mouse.sort_values('task_num').rename(columns={'dur_s': 'mouse_s'})

# Voice
df_voice = pd.read_csv('ELAN_outputs/bench_1_voice.txt', sep='\t', header=None,
                 names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_voice = df_voice[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_voice = df_voice.groupby('value')['dur_s'].mean().reset_index()
result_voice['task_num'] = result_voice['value'].str.extract(r'(\d+)').astype(int)
result_voice = result_voice.sort_values('task_num').rename(columns={'dur_s': 'voice_s'})

# Merge and calculate percentage increase
combined = result_mouse.merge(result_voice[['task_num', 'voice_s']], on='task_num')
combined['pct_increase'] = ((combined['voice_s'] - combined['mouse_s']) / combined['mouse_s']) * 100

print(combined[['value', 'mouse_s', 'voice_s', 'pct_increase']].round(2).to_string(index=False))
```

```python
import pandas as pd

# Mouse/keyboard
df_mouse = pd.read_csv('ELAN_outputs/bench_1_mouse_keyboard.txt', sep='\t', header=None,
                       names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_mouse = df_mouse[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_mouse = df_mouse.groupby('value')['dur_s'].agg(mouse_s='mean', mouse_sd='std').reset_index()
result_mouse['task_num'] = result_mouse['value'].str.extract(r'(\d+)').astype(int)
result_mouse = result_mouse.sort_values('task_num')



# Voice
df_voice = pd.read_csv('ELAN_outputs/bench_1_voice.txt', sep='\t', header=None,
                       names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_voice = df_voice[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_voice = df_voice.groupby('value')['dur_s'].agg(voice_s='mean', voice_sd='std').reset_index()
result_voice['task_num'] = result_voice['value'].str.extract(r'(\d+)').astype(int)
result_voice = result_voice.sort_values('task_num')

# Merge and calculate percentage increase
combined = result_mouse.merge(result_voice[['task_num', 'voice_s', 'voice_sd']], on='task_num')
# combined['pct_increase'] = ((combined['voice_s'] - combined['mouse_s']) / combined['mouse_s']) * 100
combined['ratio'] = (combined['voice_s'] / combined['mouse_s']).round(1)

# print(combined[['value', 'mouse_s', 'mouse_sd', 'voice_s', 'voice_sd', 'pct_increase']].round(2).to_string(index=False))
print(combined[['value', 'mouse_s', 'mouse_sd', 'voice_s', 'voice_sd', 'ratio']].round(2).to_string(index=False))
```

```python
import pandas as pd

# Mouse/keyboard
df_mouse = pd.read_csv('ELAN_outputs/bench_1_mouse_keyboard.txt', sep='\t', header=None,
                       names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_mouse = df_mouse[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_mouse = df_mouse.groupby('value')['dur_s'].agg(
    mouse_s='mean',
    mouse_sd='std',
    mouse_max='max'
).reset_index()
result_mouse['task_num'] = result_mouse['value'].str.extract(r'(\d+)').astype(int)
result_mouse = result_mouse.sort_values('task_num')

# Voice
df_voice = pd.read_csv('ELAN_outputs/bench_1_voice.txt', sep='\t', header=None,
                       names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_voice = df_voice[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_voice = df_voice.groupby('value')['dur_s'].agg(
    voice_s='mean',
    voice_sd='std',
    voice_max='max'
).reset_index()
result_voice['task_num'] = result_voice['value'].str.extract(r'(\d+)').astype(int)
result_voice = result_voice.sort_values('task_num')

# Merge and calculate ratio
combined = result_mouse.merge(
    result_voice[['task_num', 'voice_s', 'voice_sd', 'voice_max']], on='task_num'
)
combined['ratio'] = (combined['voice_s'] / combined['mouse_s']).round(1)

print(combined[['value', 'mouse_s', 'mouse_sd', 'mouse_max',
                'voice_s', 'voice_sd', 'voice_max', 'ratio']].round(2).to_string(index=False))
```

```python
import pandas as pd

# Mouse/keyboard
df_mouse = pd.read_csv('ELAN_outputs/bench_1_mouse_keyboard.txt', sep='\t', header=None,
                       names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_mouse = df_mouse[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_mouse = df_mouse.groupby('value')['dur_s'].agg(
    mouse_s='mean',
    mouse_sd='std',
    mouse_max='max'
).reset_index()
result_mouse['mouse_cv'] = (result_mouse['mouse_sd'] / result_mouse['mouse_s'] * 100).round(1)
result_mouse['task_num'] = result_mouse['value'].str.extract(r'(\d+)').astype(int)
result_mouse = result_mouse.sort_values('task_num')

# Voice
df_voice = pd.read_csv('ELAN_outputs/bench_1_voice.txt', sep='\t', header=None,
                       names=['tier', 'participant', 'begin_s', 'end_s', 'dur_s', 'value'])
df_voice = df_voice[['tier', 'begin_s', 'end_s', 'dur_s', 'value']]

result_voice = df_voice.groupby('value')['dur_s'].agg(
    voice_s='mean',
    voice_sd='std',
    voice_max='max'
).reset_index()
result_voice['voice_cv'] = (result_voice['voice_sd'] / result_voice['voice_s'] * 100).round(1)
result_voice['task_num'] = result_voice['value'].str.extract(r'(\d+)').astype(int)
result_voice = result_voice.sort_values('task_num')

# Merge and calculate ratio
combined = result_mouse.merge(
    result_voice[['task_num', 'voice_s', 'voice_sd', 'voice_cv', 'voice_max']], on='task_num'
)
combined['ratio'] = (combined['voice_s'] / combined['mouse_s']).round(1)

print(combined[['value', 
                'mouse_s', 'mouse_sd', 'mouse_cv', 'mouse_max',
                'voice_s', 'voice_sd', 'voice_cv', 'voice_max',
                'ratio']].round(2).to_string(index=False))
```
